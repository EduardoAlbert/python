## Análise 3

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
```

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
fig, ax = plt.subplots(figsize=(8, 6))
colors = ['pink', 'green', 'blue']

sns.barplot(x='fuelType', y='price', hue='gearbox', data=df, palette=colors)

ax.set_title('Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio')
ax.set_xlabel('Tipo de Combustível', fontsize=14)
ax.set_ylabel('Preço Médio', fontsize=14)

plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/fueltype-vehicleType-price.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
```

```python
fig, ax1 = plt.subplots(figsize=(8, 6))
colors1 = ['blue', 'yellow', 'gray']

sns.barplot(x='vehicleType', y='powerPS', hue='gearbox', data=df, palette=colors1)

ax1.set_title('Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio')
ax1.set_xlabel('Tipo de Veículo', fontsize=14)
ax1.set_ylabel('Potência Média', fontsize=14)

plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/vehicletype-fueltype-power.png")
```

```python
# Pôtencia por tipo de veículo
fig2, ax2 = plt.subplots(figsize=(8, 6))

g = sns.barplot(x='vehicleType', y='powerPS', data=df, ax=ax2)

ax2.set_title('Tipo de Veículo X Potência', fontsize=18)
ax2.set_xlabel('Tipo de Veículo', fontsize=14)
ax2.set_ylabel('Potência', fontsize=14)

plt.show()
```
