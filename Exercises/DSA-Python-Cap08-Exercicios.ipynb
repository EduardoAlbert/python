### Exercício 1

```python
# Crie um array NumPy com 1000000 e uma lista com 1000000.
# Multiplique cada elemento do array e da lista por 2 e calcule o tempo de execução com cada um dos objetos (use %time).
# Qual objeto oferece melhor performance, array NumPy ou lista?
```

```python
import numpy as np
```

```python
my_arr = np.arange(1000000)
my_list = list(range(1000000))
```

```python
# Array NumPy
%time for _ in range(10): my_arr2 = my_arr * 2
```

```python
# Python List
%time for _ in range(10): my_list2 = [x * 2 for x in my_list]
```

### Exercício 6

```python
# Considere os 3 arrays abaixo
# Retorne o valor do array xarr se o valor for True no array cond. Caso contrário, retorne o valor do array yarr.
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
```

```python
result = [(x if c else y) for x, y, c in zip(xarr, yarr, cond)]
result
```

### Exercício 10

```python
# Crie um banco de dados no SQLite, crie uma tabela, insira registros, 
# consulte a tabela e retorne os dados em dataframe do Pandas
```

```python
import sqlite3 as sql
```

```python
con = sql.connect('Users.db')
```

```python
cur = con.cursor()
```

```python
# Instrução para criar o banco de dados
sql_create = 'create table users' \
             '(id integer primary key,' \
             'nome varchar(50),' \
             'idade int)'
```

```python
cur.execute(sql_create)
```

```python
# Instrução para inserir dados no banco de dados
sql_insert = "insert into users values (?, ?, ?)"
```

```python
data = [(1, 'Eduardo', 18),
        (2, 'Alana', 45),
        (3, 'Henrique', 50)]

con.executemany(sql_insert, data)
```

```python
con.commit()
```

```python
# Instrução para selecionar os dados do banco de dados
sql_select = "select * from users"
```

```python
cur.execute(sql_select)
```

```python
data = cur.fetchall()
data
```

```python
# Inserindo os dados em um DataFrame
users_df = pd.DataFrame(data, columns=[x[0] for x in cur.description])

users_df
```
